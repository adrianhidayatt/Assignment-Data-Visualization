**3. Mencari nilai Rata-rata, Standar deviasi, Skewness, Median, Modus, Energi, Min, Max**

```python
# Fungsi untuk menghitung statistik yang diminta dan menyimpannya di parameter dataset baru
def hitung_statistik(data):
    statistics = pd.DataFrame(columns=['Rata-rata', 'Standar Deviasi', 'Skewness', 'Median', 'Modus', 'Energi', 'Min', 'Max'])
    for col_idx, col_name in enumerate(data.columns):
        column_data = data[col_name]
        statistics.loc[col_idx] = [
            column_data.mean(),
            column_data.std(),
            skew(column_data),
            column_data.median(),
            column_data.mode().iloc[0],
            (column_data ** 2).mean(),
            column_data.min(),
            column_data.max()
        ]
    return statistics
```

```python
stats_file = hitung_statistik(dataset)
stats_final = stats_file.T.rename(columns={col: f"Kolom {col+1}" for col in stats_file.T})
stats_final
```

**4. Plot sinyal diatas dari no. 2 dengan multiplot 2x3, sinyal yang diplot dapat dipilih dari 6 kolom secara bebas**

```python
kolom_terpilih = dataset.sample(n=6, axis=1)
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
for i, ax in enumerate(axes.flatten()):
    kolom = kolom_terpilih.columns[i]
    ax.plot(dataset.index, dataset[kolom])
    ax.set_title(f'Kolom {kolom}')
    ax.set_xlabel('Indeks')
    ax.set_ylabel('Nilai')
plt.tight_layout()
plt.show()
```

**5. Plot scatterplot dari hasil no. 3 untuk rata-rata, buat warna dan tanda yang berbeda untuk setiap 10 kolom, kolom 1-10 warna merah *, kolom 11-20 warna biru o, kolom 21-30 warna hijau v, dst (bebas). Dimana x adalah nilai rata-rata dan y adalah nilai standar deviasi.**

```python
# Menentukan warna dan marker untuk setiap kelompok 10 kolom
colors = ['red', 'blue', 'green', 'cyan', 'magenta', 'yellow']
markers = ['*', 'o', 'v', '^', 's', 'd']
# Mengatur ukuran figure
plt.figure(figsize=(10, 8))
# Plot scatterplot untuk setiap kelompok 10 kolom
for i in range(0, len(stats_final.columns), 10):
    subset = stats_final.iloc[:, i:i+10]
    color_index = i // 10 % len(colors)
    marker_index = i // 10 % len(markers)
    plt.scatter(subset.loc['Rata-rata'], subset.loc['Standar Deviasi'], color=colors[color_index], marker=markers[marker_index], label=f'Kolom {i+1}-{i+10}')
# Menambahkan label dan judul plot
plt.xlabel('Rata-rata')
plt.ylabel('Standar Deviasi')
plt.legend()
plt.title('Scatterplot Rata-rata (Mean) vs Standar Deviasi (Std Dev) dengan Warna dan Tanda Berbeda untuk Setiap Kelompok 10 Kolom')
# Menampilkan plot
plt.show()
```
